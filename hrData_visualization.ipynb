```python
import numpy as np
import pandas as pd

# txt 파일 읽기, 공백 기준으로 분류해 데이터셋 생성
data = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/star_class.txt', engine='python')

# 필요한 열 추출
df = data[['Temperature (K)', 'Luminosity (L/Lo)', 'Star category']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 산점도 그리기
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=df, x='Temperature (K)', y='Luminosity (L/Lo)', hue='Star category', palette='viridis')

# 그래프 제목 및 축 라벨 설정
scatter_plot.set_title('Hertzsprung-Russell Diagram')
scatter_plot.set_xlabel('Temperature (K)')
scatter_plot.set_ylabel('Luminosity (L/Lo)')

# Set the axis to logarithmic scale
plt.xscale('log')
plt.yscale('log')

# x축의 온도를 역순으로 설정 (온도는 왼쪽에서 오른쪽으로 감소해야 함)
scatter_plot.invert_xaxis()

# 범례 위치 조정
plt.legend(loc='upper left')

# 그래프 보여주기
plt.show()
```
